```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.datasets import make_moons, make_blobs, make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd

# Dataset 1: Non-spherical (DBSCAN-friendly)
X_moons, y_moons = make_moons(n_samples=300, noise=0.05, random_state=42)
X_moons = StandardScaler().fit_transform(X_moons)

# Dataset 2: Varying densities (DBSCAN struggles)
X_blobs, y_blobs = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)
X_blobs = StandardScaler().fit_transform(X_blobs)

# Dataset 3: Concentric Circles (Non-linearly separable, DBSCAN-friendly)
X_circles, y_circles = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)
X_circles = StandardScaler().fit_transform(X_circles)

# Plotting helper function
def plot_clusters(X, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Silhouette Score Helper Function
def evaluate_clustering(X, labels, title):
    if len(set(labels)) > 1 and -1 not in labels:
        score = silhouette_score(X, labels)
        print(f"Silhouette Score for {title}: {score:.2f}")
    else:
        print(f"Silhouette Score for {title}: Not Applicable")

# k-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_moons)
plot_clusters(X_moons, kmeans_labels, "k-Means on Moon Dataset")
evaluate_clustering(X_moons, kmeans_labels, "k-Means on Moon Dataset")

# Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters=2)
hc_labels = hc.fit_predict(X_moons)
plot_clusters(X_moons, hc_labels, "Hierarchical Clustering on Moon Dataset")
evaluate_clustering(X_moons, hc_labels, "Hierarchical Clustering on Moon Dataset")

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_moons)
plot_clusters(X_moons, dbscan_labels, "DBSCAN on Moon Dataset")
evaluate_clustering(X_moons, dbscan_labels, "DBSCAN on Moon Dataset")

# Performance on Blobs Dataset
kmeans_labels_blobs = kmeans.fit_predict(X_blobs)
plot_clusters(X_blobs, kmeans_labels_blobs, "k-Means on Blobs Dataset")
evaluate_clustering(X_blobs, kmeans_labels_blobs, "k-Means on Blobs Dataset")

hc_labels_blobs = hc.fit_predict(X_blobs)
plot_clusters(X_blobs, hc_labels_blobs, "Hierarchical Clustering on Blobs Dataset")
evaluate_clustering(X_blobs, hc_labels_blobs, "Hierarchical Clustering on Blobs Dataset")

dbscan_labels_blobs = dbscan.fit_predict(X_blobs)
plot_clusters(X_blobs, dbscan_labels_blobs, "DBSCAN on Blobs Dataset")
evaluate_clustering(X_blobs, dbscan_labels_blobs, "DBSCAN on Blobs Dataset")

# Performance on Circles Dataset
kmeans_labels_circles = kmeans.fit_predict(X_circles)
plot_clusters(X_circles, kmeans_labels_circles, "k-Means on Circles Dataset")
evaluate_clustering(X_circles, kmeans_labels_circles, "k-Means on Circles Dataset")

hc_labels_circles = hc.fit_predict(X_circles)
plot_clusters(X_circles, hc_labels_circles, "Hierarchical Clustering on Circles Dataset")
evaluate_clustering(X_circles, hc_labels_circles, "Hierarchical Clustering on Circles Dataset")

dbscan_labels_circles = dbscan.fit_predict(X_circles)
plot_clusters(X_circles, dbscan_labels_circles, "DBSCAN on Circles Dataset")
evaluate_clustering(X_circles, dbscan_labels_circles, "DBSCAN on Circles Dataset")

# Comparison Table
data = {
    'Feature': ['Definition', 'Approach', 'Number of Clusters', 'Cluster Shape', 'Initialization', 'Result', 'Interpretability', 'Strengths', 'Limitations'],
    'k-Means': ['Partitioning algorithm that assigns points to k clusters based on centroids', 'Iteratively minimizes variance within k clusters', 'Requires predefined k', 'Prefers spherical clusters', 'Randomly selects k initial centroids', 'Hard assignments—each point belongs to a single cluster', 'Moderate—cluster assignments but no hierarchy', 'Simple, fast, and efficient on large datasets', 'Sensitive to initial centroids and k choice'],
    'Hierarchical Clustering': ['Builds a hierarchy of clusters using distance metrics', 'Agglomerative (bottom-up) or divisive (top-down)', 'Can be determined from dendrogram but subjective', 'Works well with various shapes but can be unstable', 'No initialization needed', 'Hierarchical structure (tree/dendrogram)', 'High—dendrogram can be analyzed', 'Can capture hierarchical relationships', 'Computationally expensive for large datasets'],
    'DBSCAN': ['Density-based algorithm that groups points based on density and marks noise', 'Expands clusters from core points with min_samples in eps radius', 'Determined by density, no need to predefine clusters', 'Can handle arbitrarily shaped clusters', 'No initialization needed', 'Clusters and noise labels (-1)', 'Low—no cluster hierarchy', 'Detects noise and arbitrarily shaped clusters', 'Sensitive to eps and min_samples parameters, struggles with varying densities']
}
comparison_table = pd.DataFrame(data)
print(comparison_table.to_string(index=False))

print("Clustering Completed")
```
